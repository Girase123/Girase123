```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Parameters
n = 20          # number of trials
p = 0.5         # probability of success
x = np.arange(0, n+1)  # possible outcomes

# Binomial PMF
pmf = binom.pmf(x, n, p)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# Compare p = 0.3, 0.5, 0.7
plt.figure(figsize=(10, 6))
for p in [0.3, 0.5, 0.7]:
    pmf = binom.pmf(x, n, p)
    plt.plot(x, pmf, marker='o', label=f'p={p}')

plt.title(f'Binomial Distribution Comparison (n={n})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compare n = 10, 20, 50 with fixed p = 0.5
plt.figure(figsize=(10, 6))
for n in [10, 20, 50]:
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, 0.5)
    plt.plot(x, pmf, marker='o', label=f'n={n}')

plt.title('Effect of Increasing Trials on Binomial Distribution (p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
